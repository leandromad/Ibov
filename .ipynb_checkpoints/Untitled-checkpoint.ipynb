```python
import yfinance as yf
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
```

```python
tickers = pd.read_excel("composicao_ibov.xlsx")
```

```python
#pegando lista de tickers

intervalo_tempo = tickers.columns

lista_tickers = []

for mes in intervalo_tempo:

    tickers_no_ano = tickers[mes]

    tickers_no_ano = tickers_no_ano.dropna()

    tickers_no_ano = tickers_no_ano 

    lista_tickers.append(tickers_no_ano)


lista_tickers_finais = pd.concat(lista_tickers)

lista_tickers_finais = lista_tickers_finais.drop_duplicates()

lista_tickers_finais = list(lista_tickers_finais)

lista_tickers_finais

intervalo_tempo
```

```python
lista_tickers_finais
```

```python
driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()))
dict_setor = {}
for ticker in lista_tickers_finais:
    try:
        url = f"https://statusinvest.com.br/acoes/{ticker}"
        driver.get(url)
        time.sleep(3)
        setor = driver.find_element("xpath", "/html/body/main/div[5]/div[1]/div/div[3]/div/div[1]/div/div/div/a/strong")
        setor = str(setor.text)
        dict_setor[ticker] = setor
    except:
        dict_setor[ticker] = "-"
    print(dict_setor)
```

```python
valores_unicos = set(dict_setor.values())
valores_unicos
```

```python
for coluna in tickers.columns:
    tickers[coluna] = tickers[coluna].map(dict_setor)
```

```python
# contar ocorrências de valores em cada coluna
counts = tickers.apply(pd.Series.value_counts)

counts = counts.fillna(0)

counts = counts.astype(int)
counts = counts[1:]
counts
```
